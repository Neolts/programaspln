#### Numerical Analisys and Visualization
##### Medidas de Tendencia Central
Las medidas de tendencia central de un *dataset* están representadas básicamente por 
tres parámetros:
- La media artimética o promedio.
- La moda
- La desviación estandard

```python
# Obtener estadísticas por columna.
# La media es el promedio de un conjunto de datos. Se calcula sumando
# todos los valores y dividiéndolos por el número total de valores.
df["city-kpl"].mean()
```

```python
# La moda es el valor que aparece con mayor frecuencia en un conjunto de datos.
# Si no hay un valor que aparezca más veces que los demás, entonces no hay moda para este conjunto de datos.
df["city-kpl"].mode()
```

```python

# La desviación estándar es una medida de dispersión que indica cuánto varían los datos de su media.
# Primero, calculamos la media de los datos
# Luego, calculamos la diferencia entre cada valor y la media, y la elevamos al cuadrado.
#  Sumamos todas estas diferencias al cuadrado.
# Dividimos esta suma por el número total de valores 
# Finalmente, tomamos la raíz cuadrada del resultado.

df["city-kpl"].std()
```

#### Las Gráficas de Barras
**¿Qué es una gráfica de barras?**

Una gráfica de barras es una representación visual de datos que utiliza barras rectangulares para mostrar la frecuencia, el recuento o la proporción de diferentes categorías. Cada barra representa una categoría específica y su altura es proporcional a la frecuencia o valor asociado con esa categoría.

**Componentes de una gráfica de barras:**

- **Eje X (horizontal).** En una gráfica de barras, el eje X representa las categorías o grupos de datos. Cada barra se coloca a lo largo del eje X en función de su categoría correspondiente.

- **Eje Y (vertical).** El eje Y representa la frecuencia, el recuento o la proporción de cada categoría. La altura de cada barra indica la magnitud de la variable representada.

- **Barras.** Las barras son rectángulos verticales (o horizontales en un gráfico de barras horizontales) que representan visualmente la cantidad o valor asociado con cada categoría. La longitud (o altura) de cada barra es proporcional a la magnitud de la variable representada.

**Uso y aplicaciones:**

Las gráficas de barras son útiles para:

1. Comparar la frecuencia o el recuento de diferentes categorías.
2. Visualizar la distribución de datos categóricos.
3. Comparar datos entre diferentes grupos o períodos de tiempo.
4. Identificar tendencias, patrones o disparidades en los datos.
5. Tipos de gráficas de barras:

*Gráfica de barras verticales:* Las barras se colocan verticalmente a lo largo del eje X.
*Gráfica de barras horizontales:* Las barras se colocan horizontalmente a lo largo del eje Y.

**Consideraciones adicionales:**

Es importante etiquetar claramente los ejes X e Y y proporcionar una leyenda si es necesario para que la gráfica sea fácil de interpretar.
Las gráficas de barras son efectivas para datos categóricos, pero para datos numéricos continuos, es posible que sea más apropiado utilizar un histograma.

Las gráficas de barras son una herramienta versátil y efectiva para visualizar y comparar datos categóricos, lo que permite una comprensión rápida y clara de la distribución y las tendencias en los datos.

```python
# importar la librería matplotlib.pyplot
# pip install matplotlib

# Generar 5 datos aleatorios
datos = np.random.randint(1, 20, size=5)  # Genera 5 números enteros aleatorios entre 1 y 20

# Crear una lista de etiquetas para las barras
etiquetas = ['Auto 1', 'Auto 2', 'Auto 3', 'Auto 4', 'Auto 5']
# Crear la gráfica de barras
plt.bar(etiquetas, datos, color='blue')
# Agregar título y etiquetas a los ejes
plt.title('Ejemplo de Gráfica de Barras')
plt.xlabel('Automóviles')
plt.ylabel('Valores')

# Mostrar la gráfica
plt.grid(True, linestyle='--', alpha=0.7)  # Agregar una cuadrícula
plt.tight_layout()  # Ajustar diseño
plt.show()
```

```python
# Gráfica de distribución de marca de automóviles 
# Obtener el número de autos  de cada marca 
nummarcas = df["make"].value_counts()

# Graficar la distribución de marcas de automóviles en un gráfico de barras
ax = nummarcas.plot(kind="bar", figsize=(10, 6), color='skyblue', edgecolor='black')
nummarcas.plot(kind="bar", figsize=(10, 6), color='skyblue', edgecolor='black')
# Agregar título y etiquetas a los ejes
plt.title("Distribución de Marcas de Automóviles")
plt.xlabel("Marcas")
plt.ylabel("Número")
# Agregar la cantidad por marca arriba de cada barra
for i, v in enumerate(nummarcas):
   ax.text(i, v + 0.2, str(v), ha='center', va='bottom')

# Mostrar la gráfica
# Rotar las etiquetas del eje x para mayor claridad
plt.xticks(rotation=45) 
# Agregar una cuadrícula horizontal 
plt.grid(axis='y', linestyle='--', alpha=0.7) 
# Ajustar diseño 
plt.tight_layout()  
plt.show()
```

```python
# Crear una gráfica de rendimiento de cada marca en base al kilometraje
plt.figure(figsize=(12, 8))  # Aumentar el tamaño de la figura

# Agrupar los datos por marca y calcular el kilometraje promedio en ciudad y autopista
avg_mileage = df.groupby('make')[['city-kpl', 'highway-kpl']].mean()

# Graficar el rendimiento de las marcas de automóviles
avg_mileage.plot(kind='bar', width=0.8)  # Ajustar el ancho de las barras
# Datos de la gráfica
plt.title('Rendimiento de Marcas de Automóviles')
plt.xlabel('Marca')
plt.ylabel('Kilometraje Promedio (kpl)')
plt.xticks(rotation=45, ha='right')  # Rotar y ajustar la alineación de las etiquetas x
plt.legend(['Ciudad', 'Autopista'])
plt.tight_layout()
plt.show()
```

```python
# Ejemplo de correlación
# Generar dos conjuntos de datos aleatorios
np.random.seed(0)  # Establecer semilla aleatoria para reproducibilidad
datos1 = np.random.randint(1, 10, size=5)
datos2 = np.random.randint(1, 10, size=5)

# Calcular la correlación entre los dos conjuntos de datos. Pearson
correlacion = np.corrcoef(datos1, datos2)[0, 1]

# Graficar los datos en un diagrama de dispersión
plt.scatter(datos1, datos2, color='blue')
plt.title(f'Diagrama de Dispersión\nCorrelación: {correlacion:.2f}')
plt.xlabel('Datos 1')
plt.ylabel('Datos 2')
plt.grid(True, linestyle='--', alpha=1.0)
plt.show()
```
